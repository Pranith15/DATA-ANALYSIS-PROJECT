```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Bar Plot

```python
top10 = df.nlargest(10,'subscribers')
sns.barplot(x='subscribers', y='channel_type', data=top10)
plt.title('Top 10 Channels by Subscribers')
plt.show()
```

# Hist Plot

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['video views'], bins=50, kde=True)
plt.title('Distribution of Video Views')
plt.xlabel('Video Views')
plt.show()
```

# Box Plot

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='category', y='highest_monthly_earnings', data=df)
plt.xticks(rotation=90)
plt.title('Monthly Earnings by Channel Category')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='subscribers', y='video views', data=df, hue='category')
plt.title('Subscribers vs Video Views by Category')
plt.show()
```
